## ЗАДАЧА 5 - апроксимация матрицы A (Задача решена с помощью нейронки, так как практики не было, было просто видео(и то мельком) и нету понятия как работать и решать задачу включая также понимание формулировки)

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100
A = np.random.randn(n, n)

r_values = np.arange(2, n+1)
errors = []

# Полное SVD для исходной матрицы
U, S, Vt = np.linalg.svd(A, full_matrices=False)

# Визуализация разницы для выбранных рангов
selected_ranks = [10, 30, 60, 100]  # Можно выбрать интересующие ранги

import numpy as np
import matplotlib.pyplot as plt

n = 100
A = np.random.randn(n, n)

r_values = np.arange(2, n+1)
errors = []

# Полное SVD для исходной матрицы
U, S, Vt = np.linalg.svd(A, full_matrices=False)

# Выбор рангов для визуализации разницы
selected_ranks = [10, 30, 60, 100]

for r in r_values:
    S_r = np.diag(S[:r])
    U_r = U[:, :r]
    Vt_r = Vt[:r, :]
    A_tilde = U_r @ S_r @ Vt_r
    
    # Расчет ошибки
    error = np.linalg.norm(A - A_tilde, ord='fro')
    errors.append(error)
    
    # Визуализация разницы для выбранных рангов
    if r in selected_ranks:
        delta = A - A_tilde
        plt.figure(figsize=(6, 5))
        # Можно использовать log для улучшения видимости малых значений
        plt.imshow(np.log1p(np.abs(delta)), cmap='bwr', interpolation='none')
        plt.colorbar(label='log(1 + |разница|)')
        plt.title(f'Разница A - аппроксимация при ранге r={r}')
        plt.xlabel('Столбцы')
        plt.ylabel('Строки')
        plt.show()

# Построение графика зависимости ошибки
plt.figure(figsize=(8, 6))
plt.plot(r_values, errors, marker='o')
plt.xlabel('Ранк r')
plt.ylabel('Ошибка аппроксимации E(r)')
plt.title('Зависимость ошибки аппроксимации матрицы от ранга')
plt.grid(True)
plt.show()
```

## ЗАДАЧА 4 - нахождение дифф df (x) и градиента ∇f (x)<br>
- **Градиент**  
  ∇f (x) = ∥x∥2X<br>
  Это вектор, направленный вдоль самого себя по тому же направлению, но длина его умножена на длину исходного вектора.

- **Дифференциал**  
  Для любого небольшого изменения<br>
  df(x)(Δx) = ∥x∥2 ⟨x,Δx⟩<br>
- **Пример**<br>
  x = (3,4)<br>
  Тогда: ∥x∥2 = 3^2+4^2 = 5 - корень.<br>
  Градиент: ∇f(x) = 5(3,4) = (15,20).<br>
  Дифференциал: если вы слегка измените вектор, например добавив (0.1,0), то <br>
  df(x)(Δx) = 5⟨(3,4),(0.1,0)⟩ = 5(3×0.1+4×0) = 1.5. <br>
  То есть функция увеличится примерно на 1.5.<br>


## ЗАДАЧА 5 - нахождение дифф df (x) и градиента ∇f (x) - Евклидова норма<br>
- **Градиент**  
  ∇f (x) = x/∥x∥<br>
  Вектор единичной длины, указывающий в том же направлении, что и x.

- **Дифференциал**  
  Первый дифференциал (линейное приближение изменения) для небольшого смещения h<br>
  df(x)[h] = ⟨x,h⟩/∥x∥<br>
  
- **Пример**<br>
  x = (3,4)<br>
  Тогда: ∥x∥2 = 3^2+4^2 = 5 - корень.<br>
  Градиент: ∇f(x) = ∥x∥/x = (3/5,4/5).<br>
  Если мы немного переместимся, скажем на h=(0.2,−0.1), то приближённое изменение нормы будет
  df(x)[h]= (3)(0.2)+(4)(−0.1)/5 = 0.6−0.4/ = 0.04 <br>
  То есть функция увеличится примерно на 0.04.<br>
